```python
# main code
import os
from pdf2image import convert_from_path

def convert_pdf_to_images(pdf_dir, output_dir, dpi=200):
    for filename in os.listdir(pdf_dir):
        if filename.endswith(".pdf"):
            pdf_path = os.path.join(pdf_dir, filename)
            images = convert_from_path(pdf_path, dpi=dpi)
            print(pdf_path)
            
            for i, image in enumerate(images):
                output_filename = f"{filename[:-4]}-page-{i+1}.jpg"  # Create unique names
                output_path = os.path.join(output_dir, output_filename)
                image.save(output_path, "JPEG")
                print(output_path)
                print()
pdf_dir = "C:\\Users\\Rdhika\\Desktop\python_project\\inv_project\\INV"
output_dir = "C:\\Users\\Rdhika\\Desktop\\python_project\\inv_project\\pdf_to_img"
convert_pdf_to_images(pdf_dir, output_dir)
```

```python
#single pdf data extract
def get_details(text):
    lst = text.split("\n")
    pan_no = re.search("[A-Z]{5}\d{4}[A-Z]",text).group()
    mail_id=re.findall("[\w]{1,36}[@][a-z.]{8,14}",text)
    phone_no=re.search("[9][1][-]\d{10}",text).group()
    GSTIN_no = re.search("\d{2}[A-Z]{5}\d{4}[A-Z]\d[A-Z]{2}",text).group()
    customer_Account_no=re.search("[0]\d{11}",text).group()
    IFSC_Code = re.search("[A-Z]{4}\d{7}",text).group()
    for i, line in enumerate(lst):
        if re.findall("TAX",line):      
            card_holder_name = lst[i+1]           
            fathers_name = lst[i+2]
        inv_dict = {"pan no":[pan_no],
                      "email id" : [mail_id],
                      "phone no" : [phone_no],
                      "GSTIN no" : [GSTIN_no],
                      "customer Account no" : [customer_Account_no],
                    "IFSC code" : [IFSC_Code]
                    }
    return inv_dict
```

```python
# extract data from multipal images and appennd it in list
import os
import re
path=r"C:\Program Files\Tesseract-OCR\tesseract"
pytesseract.pytesseract.tesseract_cmd=path
folder_path = r"C:\\Users\\Rdhika\\Desktop\\python_project\\inv_project\\pdf_to_img"
files = os.listdir(folder_path)
lst_cust_name=[]
lst_pan = []
lst_dob = []
lst_Gst_No=[]
lst_Date=[]
lst_pan=[]
for i in files:
    image_path = os.path.join(folder_path,i)
    text = pytesseract.image_to_string(image_path)                     #image_to_string(image_path)
    cust_name=re.search("[A-Za-z.& ]+",text).group()
    Gst_No=re.search("[A-Z]*\d{1,2}[A-Z]{5}\d{4}[A-Z][0-9]+[A-Z0-9][A-Z0-9]+",text).group()
    Invoice_Date=re.findall("\d{1,2}[.-/]\d{1,2}[.-/]\d{2,4}",text)
    Pan_no=re.search("[A-Z]{5}[0-9]{4}[A-Z]",text).group()
    lst_cust_name.append(cust_name)
    lst_Gst_No.append(Gst_No)
    lst_Date.append(Invoice_Date)
    lst_pan.append(Pan_no)
```
